```python
import pandas as pd
col = ['image_id', 'attribute_id', 'isPresent']
data = pd.read_csv('test_label_attributes.txt', header = None, sep = " ", names = col)
data.head(10)
```

```python
col1 = ['attribute_id', 'attribute_name']
attr = pd.read_csv('attributes.txt', header =None, sep =' ',  names = col1)
attr.head(10)
```

```python
joined = data.join(attr.set_index('attribute_id'), on='attribute_id', validate='m:1')
columns_titles = ["image_id","attribute_id", "attribute_name", "isPresent"]
joined=joined.reindex(columns=columns_titles)
joined
```
